```python
import numpy as np
from copy import deepcopy
```

Import the file and load the stacks and isntructions into lists of lists

```python
with open('input.txt', 'r') as f:
    lines = f.read().splitlines()
```

```python
only_char_arr = np.array([list(line)[1::4] for line in lines[:8]]).T
original_stacks = [[c for c in stack if c != ' '][::-1] for stack in only_char_arr]
```

Read the instructions

```python
# Read isntructions and load into a list of lists with [[nrcrates, src, dst], ...]
original_procedure = [list(np.array(line.split(' '))[[1,3,5]].astype(int)) for line in lines[10:]]
```

Generated by ChatGPT and modified by as it misinterpreted the first number of the isntructions.
It thought it was the the crate number that had to be moved. While the correct interpretation is
that it is the number of crates that have to be moved, in secuential order from top to bottom.


```python
def move_crates_part1(stacks, procedure):
    stacks_tmp = deepcopy(stacks)
    # Iterate over the rearrangement procedure and apply each
    # step to the stacks of crates
    for step in procedure:
        # Get the number of crates to move, the source stack, and the
        # destination stack
        nrcrates, src, dst = step

        # Find the index of the source and destination stacks
        # in the list of stacks
        src_index = src - 1
        dst_index = dst - 1

        # Loop over the number of creates that have to be moved
        for i in range(nrcrates):
            # Remove the crate from the source stack and add it to
            # the destination stack
            stacks_tmp[dst_index].append(stacks_tmp[src_index].pop())

    return stacks_tmp
```

```python
final_stack_part1 = move_crates_part1(original_stacks, original_procedure)
final_stack_part1
```

```python
solution_part1 = ''
for stack in final_stack_part1:
    solution_part1 += stack[-1]
solution_part1
```

# Part 2

```python
def move_crates_part2(stacks, procedure):
    stacks_tmp = deepcopy(stacks)
    # Iterate over the rearrangement procedure and apply each
    # step to the stacks of crates
    for step in procedure:
        # Get the number of crates to move, the source stack, and the
        # destination stack
        nrcrates, src, dst = step

        # Find the index of the source and destination stacks
        # in the list of stacks
        src_index = src - 1
        dst_index = dst - 1

        # Remove the crate from the source stack and add it to
        # the destination stack
        stacks_tmp[dst_index] += stacks_tmp[src_index][-nrcrates:]
        stacks_tmp[src_index] = stacks_tmp[src_index][:-nrcrates]
    
    return stacks_tmp
```
